## Атрибуты:
1. price - Цена продажи дома.
2. lotsize - Площадь участка недвижимости в квадратных футах.
3. bedrooms - Количество спален.
4. bathrooms - Количество полностью оборудованных ванных комнат.
5. stories - Количество этажей без учета подвала.
6. driveway - Есть ли у дома подъездная дорожка
7. recreation - Есть ли в доме комната отдыха
8. fullbase - Есть ли в доме полностью законченный подвал
9. gasheat - Используется ли в доме газ для нагрева горячей воды
10. aircon - Есть ли центральное кондиционирование воздуха
11. garage - Количество мест в гараже.
12. prefer - Расположен ли дом в предпочтительном районе города

```python
#описательная статистика числовых данных
df.describe()
```

Mетод DataFrame.describe() для числовых данных включает: count - количество значений, mean - среднее значение, std - среднеквадратическое отклонение, min - минимальное значение, max - максимальное значение, а также нижний (25), 50 и верхний (75) процентили.

```python
#описательная статистика данных типа object
df.describe(include=[object])
```

Mетод DataFrame.describe() для данных типа object включает: count - количество значений, unique - количество уникальных значений, top - наиболее распространённое значение и freq - частота наиболее распространенного значения.
